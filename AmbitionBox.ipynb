```python
import pandas as pd
import numpy as np
import requests
from bs4 import BeautifulSoup
import warnings
warnings.filterwarnings('ignore')
```

```python
url = "https://www.ambitionbox.com/list-of-companies?campaign=desktop_nav"
header = {"User-Agent":"Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/110.0.0.0 Safari/537.36"}
```

```python
webpage = requests.get(url,headers=header).text
```

```python
soup = BeautifulSoup(webpage,'html.parser')
```

```python
company = soup.find_all('div',class_="company-content-wrapper")
```

```python
Name = []
Rating = []
Review = []
Type = []
Location = []
Years_Old = []
Employees = []


for i in company:
    Name.append(i.find('h2').text.strip())
    Rating.append(i.find('p',class_="rating").text.strip())
    Review.append(i.find('a',class_="review-count").text.strip())
    Type.append(i.find_all('p',class_="infoEntity sbold-list-header")[0].text.strip())
    Location.append(i.find_all('p',class_="infoEntity sbold-list-header")[1].text.strip())
    Years_Old.append(i.find_all('p',class_="infoEntity sbold-list-header")[2].text.strip())
    try:
        Employees.append(i.find_all('p',class_="infoEntity sbold-list-header")[3].text.strip())
    except:
        Employees.append(np.nan)

Data = {'Company_Name': Name,
        'Company_Rating':Rating,
        'Company_Review':Review,
        'Company_Type':Type,
        'Company_Location':Location,
        'Company_Years_Old':Years_Old,
        'Company_Employees':Employees }

df = pd.DataFrame(Data)

df.to_csv('Data.csv',index=False)

df
```

```python
final = pd.DataFrame()

for j in range(1,334):
  Url = "https://www.ambitionbox.com/list-of-companies?page={}".format(j)

  header = {"User-Agent":"Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/110.0.0.0 Safari/537.36"}
  response = requests.get(Url,headers=header).text
  soup = BeautifulSoup(response, "html.parser")
  company = soup.find_all('div',class_="company-content-wrapper")

  Name = []
  Rating = []
  Review = []
  Type = []
  Head_Quarter = []
  Old = []
  Employees = []

  for i in company:
    try:
       Name.append(i.find('h2').text.strip())
    except:
       Name.append(np.nan)

    try:
       Rating.append(i.find('p',class_='rating').text.strip())
    except:
       Rating.append(np.nan)
   
    try:
      Review.append(i.find('a' , class_='review-count').text.strip())
    except:
      Review.append(np.nan)

    try:
      Type.append(i.find_all('p',class_='infoEntity')[0].text.strip())
    except:
      Type.append(np.nan)

    try:
      Head_Quarter.append(i.find_all('p',class_='infoEntity')[1].text.strip())
    except:
      Head_Quarter.append(np.nan)
    
    try:
      Old.append(i.find_all('p',class_='infoEntity')[2].text.strip())
    except:
      Old.append(np.nan)

    try:
      Employees.append(i.find_all('p',class_='infoEntity')[3].text.strip())
    except:
      Employees.append(np.nan)

    data = {'Company_Name':Name,
          'Company_Rating':Rating,
          'Company_Review':Review,
          'Company_Type':Type,
          'Company_Head_Quarter':Head_Quarter,
          'Company_Old':Old,
          'Company_Employees':Employees }

    df = pd.DataFrame(data)

    final=final.append(df)
```

```python
final.to_csv("amb.csv")
```
